### Bercerita dengan Python

Mari kita coba!

````python
name = input('Enter file: (masukkan intro.txt sebagai demonstrasi) ')
handle = open(name, 'r')
counts = dict()

for line in handle:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word, 0) + 1

bigcount = None
bigword = None
for word, count in list(counts.items()):
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count

print('kata paling sering diulang adalah:', '\"'+bigword+'\"', 'sebanyak', bigcount, 'kali')
````

#### Kombinasi boolean

and --> Semua kondisi True maka True. Apabila ada satu kondisi False, maka akan False

or --> Salah satu kondisi True, maka True. Apabila semua kondisi False, maka False

not --> Membalikkan nilai True/False

Urutan Python membaca adalah not - and - or

````python
print(True and True)  #kedua nilai True, maka True
print(True and False)  #salah satu nilai False, maka False
print(True or False)  #salah satu nilai True, maka True
print(True and False or False)  #dibaca (True and False) or False, karena tidak ada nilai True, maka False
print(True or False and True)  #dibaca True or (False and True), karena salah satu nilai True, maka True
print(False or not False) #dibaca False or (not False), maka True
print(True and False or not False)  #dibaca ((True and False) or (not False)), maka True
````

#### ðŸ’¡ **Indexing/Slicing**

Setiap karakter dalam string memiliki indeks
```
Positive Index   0   1   2   3   4   5
                 P   Y   T   H   O   N
Negative Index  -6  -5  -4  -3  -2  -1 
```
dengan indeks tersebut, string dapat "dipotong". Hasil slicing juga merupakan string.

```string[start:finish:leap(optional)]```

````python
# Contoh penerapan slicing
kata = 'PYTHON'

# Mengambil huruf pertama
print(kata[0])

# Mengambil tiga huruf pertama (perhatikan pemakaian finishnya)
print(kata[:3])

# Mengambil huruf kedua sampai dengan keempat (perhatikan start dan finish nya)
print(kata[1:4])

# Menerapkan leap untuk mengambil huruf kedua, keempat, dst.
print(kata[1::2])

# Membalik urutan huruf (perhatikan penggunaan leap negatif)
print(kata[::-1])

# Mengambil 3 huruf dibelakang dan membaliknya
print(kata[-1:-4:-1])
````

### Lambda

Pada python, kita dapat menulis function sederhana menggunakan lambda, dengan lambda function dapat dijalankan tanpa mendefinisikan function nya terlebih dahulu (anonymous function).

contoh Lambda:

```python
# Contoh Lambda Function
fx = lambda x: x**2 + 5*x + 6

# Contoh Lambda Function dengan dua variabel
luasSegitiga = lambda alas, tinggi: alas * tinggi / 2

```

ðŸ’¡ Lambda akan sangat powerfull ketika digunakan bersama dengan map() untuk menghasilkan iterable baru

````python
# Membuat list dengan menggunakan map() dan lambda

radius = [1,2,3]
luasLingkaran = list(map(lambda r : 3.14 * r ** 2, radius))
print(luasLingkaran)
````

````python
# Membuat list dengan menggunakan map() dan lambda dengan dua variabel
panjang = [4,5,6]
lebar = [1,3,5]
luasPersegi = list(map(lambda p, l : p * l, panjang, lebar))
print(luasPersegi)
````

### If Statement

Digunakan untuk mengendalikan kode yang akan dieksekusi dengan memperhatikan pemenuhan terhadap kondisi tertentu.

Syntax:

```python

if condition1:
    # baris dieksekusi jika condition1 terpenuhi

    # baris dieksekusi jika condition2 terpenuhi
else:
    # baris dieksekusi jika condition1 & condition2 tidak terpenuhi


```

Jika kode kita membutuhkan logika percabangan sederhana, kita bisa menggunakan **ternary if**. Contoh syntaxnya sebagai berikut:

```python

```

````python
x = 15
if x % 4 == 0:
    print('kelipatan 4')
else:
    print('bukan kelipatan 4')
print('Sekian')
````

````python
# contoh conditional: menentukan tahun kabisat
def kabisat(year):
    if year % 4 == 0:            # Logic Test
        return True              # Condition1
    else:
        return False             # Condition2
````

````python
# Contoh iterary if dengan referensi ke function lainnya yang telah didefinisikan sebelumnya.
def jumlahhari():
    year = input('Masukkan tahun (yyyy):')     # mengambil input dari user, ingat tipenya otomatis adalah String
    return 366 if kabisat(int(year)) else 365  # year dipaksa untuk menjadi integer agar dapat dilakukan operasi matematis

jumlahhari()
````

### For loop (Definite Loop)

Digunakan untuk mengeksekusi sebuah blok kode secara berulang berdasarkan sebuah iterable

Syntax:

```python
for i in iterable:
    # baris dieksekusi pada setiap loop
```

ðŸ’¡ iterable dapat berupa list, tuple, set, dictionary, string, maupun range.

ðŸ’¡ range merupakan fungsi bawaan python untuk menghasilkan urutan angka secara otomatis:

format : `range(start, finish, leap)` seperti pada teknik "slicing"

````python
# Contoh for loop penggunaan range:
for i in range(1,11,2):               # start dari 1, berakhir sebelum 11, setiap 2 angka
    print(i)                          # print setiap nilai i
print('Demikian')                     # setelah habis, print Demikian
````

````python
# Contoh for: melakukan iterasi
names = ["Alice", "Bob", "Charlie"]

for name in names:
    print("Hello, " + name + "!")
````

### While Loop (Indefinite Loop)

Digunakan untuk mengeksekusi sebuah blok kode secara berulang berdasarkan sebuah kondisi tertentu yang ditetapkan selama kondisi terpenuhi.

Syntax:
```python
while condition:
    # baris dieksekusi jika condtion == True    
```

âš ï¸ __WARNING!__ kesalahan dalam membuat kode while dapat membuat loop berjalan terus-menerus

````python
# Contoh while loop

i = 0
while i < 5:
    print(i)
    i += 1 # baris ini diperlukan, jika tidak akan berisiko loop
````

````python
# Penerapan while Loop: mencari kelipatan persekutuan terkecil (KPK) dari dua bilangan

def KPK(x, y):
    i = max(x, y)
    while i % x or i % y:
        i += 1 # baris ini diperlukan, jika tidak akan berisiko loop
    return i

KPK(12, 25)
````

### Break and Continue

- **`Break`** digunakan untuk **menghentikan** proses iterasi
- **`Continue`** digunakan untuk **melompat** ke iterasi selanjutnya

````python
while True:
    x = input('Kamu siapa?')
    if x == 'exit':
        break
    print('Halo,', x)
print('Selesai')
````

````python
for i in range(1, 10):
    if i == 5:
        break
    else:
        print(i)
````

````python
for i in range(1, 10):
    if i % 3 == 0:
        continue
    else:
        print(i)
````

````python
while True:
    x = input('Kamu siapa?')
    if x[0] == '#':
        continue
    if x == 'exit':
        break
    print('Halo,', x)
print('Selesai')
````

````python
# Penerapan break dan continue: menemukan Faktor Persekutuan Terbesar
def FPB(x, y):
    for i in range(min(x,y), 0, -1):
        if x % i or y % i:
            continue
        else:
            return i
            break
        

a, b = 32, 42
FPB(a,b)
````
